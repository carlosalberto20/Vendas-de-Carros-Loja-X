Importar bibliotecas

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

Lendo dados

```python
df = pd.read_csv('empresa_x.csv')
df.head()
```

Tipos dos dados

```python
df.dtypes
```

Convertendo de string para data

```python
df['mes'] = pd.to_datetime(df['mes'])
df.dtypes
```

# Visualizar dados

```python
def grafico_linha(titulo, xlabel,ylabel, x, y, dados):
    ax = sns.lineplot(x = x, y = y, data = dados)
    sns.set_palette('Accent')
    sns.set_style('darkgrid')
    ax.figure.set_size_inches(10, 6)
    ax.set_title(titulo, loc = 'center', fontsize = 14)
    ax.set_xlabel(xlabel, fontsize = 12)
    ax.set_ylabel(ylabel, fontsize = 12)
    ax = ax    
```

```python
grafico_linha('Vendas de carros de 2017 a 2018', 'Tempo', 'Vendas (R$)',
                'mes', 'vendas', df)
```

# Aumento e aceleração das vendas

```python
#Diferença da venda do mês atual e o mês anterior

df['aumento'] = df['vendas'].diff()
df.head()
```

```python
grafico_linha('Aumento das vendas de 2017 a 2018', 'Tempo', 'Aumento',
                'mes', 'aumento', df)
```

```python
#Diferença do aumento da venda do mês atual e o mês anterior


df['aceleracao'] = df['aumento'].diff()
df.head()
```

```python
grafico_linha('Aceleração das vendas de 2017 a 2018', 'Tempo', 'Aceleração',
                'mes', 'aceleracao', df)
```
